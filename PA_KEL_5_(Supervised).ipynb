```python
# Hitung jumlah kelas M (malignant) dan B (benign)
count_class = df['diagnosis'].value_counts()

# Hitung probabilitas masing-masing kelas
prob_M = count_class['M'] / df.shape[0]
prob_B = count_class['B'] / df.shape[0]

# Ambil kolom radius_mean dan texture_mean
features = ['radius_mean', 'texture_mean']
X = df[features]

# Contoh data yang akan diprediksi
data = input("\nMasukkan data: ")
data = data.split(',')
data = [float(i) for i in data]

# Hitung nilai probabilitas untuk kelas M
likelihood_M = 1
for i in range(len(features)):
    feature_mean = df[df['diagnosis'] == 'M'][features[i]].mean()
    feature_std = df[df['diagnosis'] == 'M'][features[i]].std()
    likelihood_M *= math.exp(-((data[i]-feature_mean)**2)/(2*(feature_std**2))) / (math.sqrt(2*math.pi)*feature_std)
prob_data_M = likelihood_M * prob_M

# Hitung nilai probabilitas untuk kelas B
likelihood_B = 1
for i in range(len(features)):
    feature_mean = df[df['diagnosis'] == 'B'][features[i]].mean()
    feature_std = df[df['diagnosis'] == 'B'][features[i]].std()
    likelihood_B *= math.exp(-((data[i]-feature_mean)**2)/(2*(feature_std**2))) / (math.sqrt(2*math.pi)*feature_std)
prob_data_B = likelihood_B * prob_B

# Prediksi kelas data
if prob_data_M > prob_data_B:
    print("Hasil prediksi: malignant", prob_data_M)
else:
    print("Hasil prediksi: Benign", prob_data_B)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score

# Load the dataset
df = pd.read_csv('Breast Cancer Wisconsin (Diagnostic).csv')

# Preprocessing
X = df.iloc[:, 2:32].values
y = df.iloc[:, 1].values
y = np.where(y == 'M', 1, 0) # Convert label to binary (1 for Malignant, 0 for Benign)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# C4.5 Algorithm
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X_train, y_train)
y_pred_c45 = clf.predict(X_test)

# K-Nearest Neighbors Algorithm
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

# Naive Bayes Algorithm
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_nb = gnb.predict(X_test)

# Confusion Matrix and ROC Curve for C4.5 Algorithm
cm_c45 = confusion_matrix(y_test, y_pred_c45)
fpr_c45, tpr_c45, thresholds_c45 = roc_curve(y_test, clf.predict_proba(X_test)[:,1])
roc_auc_c45 = roc_auc_score(y_test, y_pred_c45)

# Confusion Matrix and ROC Curve for K-Nearest Neighbors Algorithm
cm_knn = confusion_matrix(y_test, y_pred_knn)
fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_test, knn.predict_proba(X_test)[:,1])
roc_auc_knn = roc_auc_score(y_test, y_pred_knn)

# Confusion Matrix and ROC Curve for Naive Bayes Algorithm
cm_nb = confusion_matrix(y_test, y_pred_nb)
fpr_nb, tpr_nb, thresholds_nb = roc_curve(y_test, gnb.predict_proba(X_test)[:,1])
roc_auc_nb = roc_auc_score(y_test, y_pred_nb)

# Plot the Confusion Matrix and ROC Curve for all Algorithms
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15,10))

# Confusion Matrix for C4.5 Algorithm
ax[0,0].imshow(cm_c45, cmap=plt.cm.Blues)
ax[0,0].set_title('C4.5 Confusion Matrix')
ax[0,0].set_xlabel('Predicted Labels')
ax[0,0].set_ylabel('True Labels')
for i in range(2):
    for j in range(2):
        ax[0,0].text(j, i, cm_c45[i,j], ha="center", va="center", color="white")

# Confusion Matrix for K-Nearest Neighbors Algorithm
ax[0,1].imshow(cm_knn, cmap=plt.cm.Blues)
ax[0,1].set_title('K-Nearest Neighbors Confusion Matrix')
ax[0,1].set_xlabel('Predicted Labels')
ax[0,1].set_ylabel('True Labels')
for i in range(2):
    for j in range(2):
        ax[0,1].text(j, i, cm_knn[i,j], ha="center", va="center", color="white")

# Confusion Matrix for Naive Bayes Algorithm
ax[0,2].imshow(cm_nb, cmap=plt.cm.Blues)
ax[0,2].set_title('Naive Bayes Confusion Matrix')
ax[0,2].set_xlabel('Predicted Labels')
ax[0,2].set_ylabel('True Labels')
for i in range(2):
    for j in range(2):
        ax[0,2].text(j, i, cm_nb[i,j], ha="center", va="center", color="white")

# ROC Curve for C4.5 Algorithm
ax[1,0].plot(fpr_c45, tpr_c45, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc_c45)
ax[1,0].plot([0, 1], [0, 1], color='navy', linestyle='--')
ax[1,0].set_xlim([0.0, 1.0])
ax[1,0].set_ylim([0.0, 1.05])
ax[1,0].set_xlabel('False Positive Rate')
ax[1,0].set_ylabel('True Positive Rate')
ax[1,0].set_title('C4.5 ROC Curve')
ax[1,0].legend(loc="lower right")

# ROC Curve for K-Nearest Neighbors Algorithm
ax[1,1].plot(fpr_knn, tpr_knn, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc_knn)
ax[1,1].plot([0, 1], [0, 1], color='navy', linestyle='--')
ax[1,1].set_xlim([0.0, 1.0])
ax[1,1].set_ylim([0.0, 1.05])
ax[1,1].set_xlabel('False Positive Rate')
ax[1,1].set_ylabel('True Positive Rate')
ax[1,1].set_title('K-Nearest Neighbors ROC Curve')
ax[1,1].legend(loc="lower right")

# ROC Curve for Naive Bayes Algorithm
ax[1,2].plot(fpr_nb, tpr_nb, color='darkorange', label='ROC curve (area = %0.2f)' % roc_auc_nb)
ax[1,2].plot([0, 1], [0, 1], color='navy', linestyle='--')
ax[1,2].set_xlim([0.0, 1.0])
ax[1,2].set_ylim([0.0, 1.05])
ax[1,2].set_xlabel('False Positive Rate')
ax[1,2].set_ylabel('True Positive Rate')
ax[1,2].set_title('Naive Bayes ROC Curve')
ax[1,2].legend(loc="lower right")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, roc_auc_score

# Load dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split dataset into training and testing set
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train models
knn = KNeighborsClassifier(n_neighbors=5)
gnb = GaussianNB()
dtc = DecisionTreeClassifier(random_state=42)

knn.fit(x_train, y_train)
gnb.fit(x_train, y_train)
dtc.fit(x_train, y_train)

# Predict
y_pred_knn = knn.predict(x_test)
y_pred_gnb = gnb.predict(x_test)
y_pred_dtc = dtc.predict(x_test)

# Evaluate
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_gnb = confusion_matrix(y_test, y_pred_gnb)
cm_dtc = confusion_matrix(y_test, y_pred_dtc)

def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: green' if v and i != 'Algorithm' else '' for i, v in zip(s.index, is_max)]

roc_auc_knn = roc_auc_score(y_test, knn.predict(x_test))
roc_auc_gnb = roc_auc_score(y_test, gnb.predict(x_test))
roc_auc_dtc = roc_auc_score(y_test, dtc.predict(x_test))

eval_knn = pd.DataFrame({
    'Algorithm': ['K-Nearest Neighbors'],
    'Accuracy (%)': [np.sum(cm_knn.diagonal())/np.sum(cm_knn)],
    'Precision (%)': [cm_knn[1,1]/np.sum(cm_knn[:,1])],
    'Recall (%)': [cm_knn[1,1]/np.sum(cm_knn[1,:])],
    'F1 Score (%)': [2*cm_knn[1,1]/(2*cm_knn[1,1]+cm_knn[0,1]+cm_knn[1,0])],
    'ROC-AUC' : [roc_auc_knn]
})

eval_gnb = pd.DataFrame({
    'Algorithm': ['Gaussian Naive Bayes'],
    'Accuracy (%)': [np.sum(cm_gnb.diagonal())/np.sum(cm_gnb)],
    'Precision (%)': [cm_gnb[1,1]/np.sum(cm_gnb[:,1])],
    'Recall (%)': [cm_gnb[1,1]/np.sum(cm_gnb[1,:])],
    'F1 Score (%)': [2*cm_gnb[1,1]/(2*cm_gnb[1,1]+cm_gnb[0,1]+cm_gnb[1,0])],
    'ROC-AUC' : [roc_auc_gnb]
})

eval_dtc = pd.DataFrame({
    'Algorithm': ['Decision Tree Classifier'],
    'Accuracy (%)': [np.sum(cm_dtc.diagonal())/np.sum(cm_dtc)],
    'Precision (%)': [cm_dtc[1,1]/np.sum(cm_dtc[:,1])],
    'Recall (%)': [cm_dtc[1,1]/np.sum(cm_dtc[1,:])],
    'F1 Score (%)': [2*cm_dtc[1,1]/(2*cm_dtc[1,1]+cm_dtc[0,1]+cm_dtc[1,0])],
    'ROC-AUC' : [roc_auc_dtc]
})

# Merge evaluation results
eval_results = pd.concat([eval_knn, eval_gnb, eval_dtc], ignore_index=True)
eval_results = eval_results.style.apply(highlight_max, subset=['Accuracy (%)', 'Precision (%)', 'Recall (%)', 'F1 Score (%)', 'ROC-AUC'])

# Display evaluation results
eval_results
```
